```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic Dataset (1).csv')
df.head()
```

```python
max_age=df['Age'].max()
print(f"Maximum Age: {max_age}")
min_age=df['Age'].min()
print(f"Minimum Age: {min_age}")
```

```python
bins=[0,15,30,45,60,75]
df['binned_age']=pd.cut(df['Age'],bins)
print(df[['binned_age','Age']].head())
```

```python
age_labels=['Young,','Young-Adults','Middle Aged','Middle-older_Age','senior']
df['binned_age']=pd.cut(df['Age'],bins,labels=age_labels)

df['binned_age'].value_counts().plot(kind='bar')
plt.title("Age Distribution")
plt.xlabel("Age Group")
plt.ylabel("count")
```

```python
labels=['PassengerId','Survived','Pclass','SibSp','Parch','Fare']

for label in labels:
  print(f"distibrution of {label}")
  sns.displot(df[label],kde=True)
  plt.show()
  print(f"Skewness: {df[label].skew()}")
```
